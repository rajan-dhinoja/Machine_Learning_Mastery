```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Vector Tutorial...

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
vector_1 = np.array([1,2,3])
vector_2 = np.array([4,5,6])

print("Vector 1: ",vector_1, "\n")
print("Vector 2: ",vector_2, "\n")
```

## Vector Arithmetic:-

```python
print(vector_1 + vector_2, "\n")
print(vector_1 - vector_2, "\n")
print(vector_1 * vector_2, "\n")
print(vector_1 / vector_2, "\n")
```

## Vector Product

```python
print(np.dot(vector_1,vector_2), "\n")
print(np.cross(vector_1,vector_2), "\n")
```
